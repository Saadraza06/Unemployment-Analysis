```python
duplicate_year = df['Year'].duplicated().any()
print(duplicate_year)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharex=True)
# Plot 1
ax[0].plot(df['Year'], df['Unemployment Rate (%)'], color='blue')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Unemployment Rate (%)')
ax[0].set_title('Unemployment Over Time')
# Plot 2
ax[1].plot(df['Year'], df['Poverty Headcount Ratio (%)'], color='red')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Poverty Headcount Ratio (%)')
ax[1].set_title('Poverty Over Time')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharex=True)
# Plot 3
ax[0].plot(df['Year'],df['GDP Growth Rate (%)'],color ='Black')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('GDP Growth Rate (%)')
ax[0].set_title('GDP Growth Rate Over Time')
# Plot 4
ax[1].plot(df['Year'],df['Government Social Spending (% of GDP)'],color='green')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Government Social Spending (% of GDP)')
ax[1].set_title('Government Social Spending (% of GDP) Over Time')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Year'],df['Unemployment Rate (%)'],color='Chocolate',label='Unemployment Rate')
plt.plot(df['Year'],df['Poverty Headcount Ratio (%)'],color='Coral',label='Poverty Headcount Ratio ')
plt.plot(df['Year'],df['Agriculture Growth Rate (%)'],color='Cyan',label='Agriculture Growth Rate')
plt.plot(df['Year'],df['Government Social Spending (% of GDP)'],color='DarkGray',label='Government Social Spending')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title("Comparison")
plt.legend()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,5))
ax.plot(df['Year'],df['Unemployment Rate (%)'],color='DarkOrange',label='Unemployment Rate',linewidth=2)
ax.plot(df['Year'],df['Poverty Headcount Ratio (%)'],color='DarkRed',label='Poverty Headcount Ratio',linewidth=2)
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment & Poverty')
ax.tick_params(axis='y', labelcolor='white')
ax.set_title('Impact of 2010 Floods')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Unemployment across the years")
plt.bar(df['Year'],df['Unemployment Rate (%)'] ,color="Magenta")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Unemployment Rate (%)'],bins=10,color='red',edgecolor='black')
plt.xlabel('Unemployment')
plt.title("Distribution of unemployment")
plt
```

```python
before_covid=df[df["Year"]<2020]
print("Average Unemployment percentage before 2020: ",before_covid["Unemployment Rate (%)"].mean(),"%")
after_covid=df[df["Year"]>=2020]
print("Average Unemployment percentage After 2020: ",after_covid["Unemployment Rate (%)"].mean(),"%")
plt.figure(figsize=(8, 5))
plt.title("Unemployment Before and after covid")
labels = ['Before COVID', 'After COVID']
values = [
    before_covid["Unemployment Rate (%)"].mean(),
    after_covid["Unemployment Rate (%)"].mean()
]
plt.bar(labels,values,color="LawnGreen", edgecolor='blue')
plt.ylabel=("Percentage")
plt.show()
```
